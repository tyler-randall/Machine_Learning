```python
df = pd.read_csv('population_vs_day.csv')

# Visualize the dataset
fig = plt.figure(figsize=(6,5))
ax1 = plt.subplot(1,1,1)
x = df["day"]
y = df["population"]
sc1 = plt.scatter(x, y, c="blue", edgecolor="blue", label="Population vs Day", s=x/1000)
plt.title("Population Growth")
plt.xlabel("Day")
plt.ylabel("Population")
plt.legend(loc="upper left")
plt.show()
```

```python
# Find minimum and maximum population
print("Minimum population:", np.min(df["population"]))
print("Maximum population:", np.max(df["population"]))

# Get training features (day)
X = df[["day"]]

# Get training target (population)
y = df["population"]
```

```python
# Train Linear Regression model
l_model = LinearRegression()
l_model.fit(X, y)
```

```python

# Train K-Nearest Neighbors model
k_model = KNeighborsRegressor()
k_model.fit(X, y)
```

```python
# Generate test data
X_test = np.arange(0, 1100, 100)
X_test = X_test.reshape(11, 1)
```

```python
# Linear model predictions
lm = l_model.predict(X_test)
```

```python
# KNN model predictions
knn = k_model.predict(X_test)
```

```python
# Visualize results
fig = plt.figure(figsize=(6,5))
ax1 = plt.subplot(1,1,1)
x = df["day"]
y = df["population"]
sc1 = plt.scatter(x, y, c="blue", edgecolor="blue", label="Original Data", s=x/1000)

# Scatter plot for Linear model
sc2 = plt.scatter(X_test, lm, color='black', label='Linear model')

# Scatter plot for KNN model
sc3 = plt.scatter(X_test, knn, color='red', marker='s', label='KNN model')

plt.xlabel("Day")
plt.ylabel("Population")
plt.legend(loc="upper right")
plt.show()
```
